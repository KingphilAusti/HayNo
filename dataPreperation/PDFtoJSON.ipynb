```python
import PyPDF2
import os

directory = "../Examples/pdfs"
files = os.listdir(directory)
pdfs = []

for file in files:
    file_path = os.path.join(directory, file)
    if os.path.isfile(file_path):
        # Process the file here
        print(file_path)
        pdfs.append(file_path)

def extract_pdf_content(file_path):
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        num_pages = len(pdf_reader.pages)

        content = []
        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            text = page.extract_text()
            content.append(text)

        return content

# for file_path in pdfs:
    # pdf_content = extract_pdf_content(file_path)

pdf_content = extract_pdf_content(pdfs[0])
```

```python
import json

# Convert the pdf_content list to a dictionary
pdf_dict = {f"element_{i}": element for i, element in enumerate(pdf_content)}

# Convert the dictionary to JSON
pdf_json = json.dumps(pdf_dict)

# Specify the file path
output_file = "output.json"

# Save the JSON to the file
with open(output_file, 'w') as file:
    json.dump(pdf_json, file)

```
